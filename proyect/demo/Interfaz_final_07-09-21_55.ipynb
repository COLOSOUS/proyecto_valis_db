```python
try:
    import tkinter as tk                # python 3
    from tkinter import font as tkfont  # python 3
except ImportError:
    import Tkinter as tk     # python 2
    import tkFont as tkfont  # python 
from tkinter import ttk
from tkinter import messagebox
from PIL import Image
import pyodbc 
import time
text_sql=""
list_sql=[]
conn=None
con=[]
cursor=None
new_indice=0
class SampleApp(tk.Tk):

    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)

        self.title_font = tkfont.Font(family='IA', size=18, weight="bold", slant="italic")
        self.configure(width=600, height=800)
        # the container is where we'll stack a bunch of frames
        # on top of each other, then the one we want visible
        # will be raised above the others
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)

        self.frames = {}
        
        for F in (StartPage, PageOne, PageTwo):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame

            # put all of the pages in the same location;
            # the one on the top of the stacking order
            # will be the one that is visible.
            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame("StartPage")

    def show_frame(self, page_name):
        '''Show a frame for the given page name'''
        frame = self.frames[page_name]
        frame.tkraise()


class StartPage(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.master = parent
        bottomframe0 = tk.Frame(self,parent)
        bottomframe0.pack(side="bottom" )
        bottomframe = tk.Frame(bottomframe0,parent)
        bottomframe.pack(side="bottom" )
        bottomframe1 = tk.Frame(bottomframe,parent)
        bottomframe1.pack(side="bottom" )
        bottomframe12 = tk.Frame(bottomframe1,parent)
        bottomframe12.pack(side="bottom" )
        bottomframe2 = tk.Frame(bottomframe12,parent)
        bottomframe2.pack(side="bottom" )
        bottomframe3 = tk.Frame(bottomframe2,parent)
        bottomframe3.pack(side="bottom" )
        
        #self.place(relwidth=1, relheight=1)
        
        
        label = tk.Label(self, text="Pagina de inicio", font=controller.title_font)
        label.pack(side="top", fill="x", pady=10)
        
        text = tk.Text(self)        
        msg = tk.Label(bottomframe0, text="Nombre servidor/IP:")
        msg.place(x=20, y=10)
        msg.pack(side="left", fill="x", pady=10)
        self.entry_var1 = tk.StringVar()
        self.input1=tk.Entry(bottomframe0, textvariable=self.entry_var1)
        self.input1.place(x=115, y=10)
        self.input1.pack(side="left", fill="x", pady=10)
        
        text = tk.Text(self)        
        msg = tk.Label(bottomframe, text="Nombre DB:")
        msg.place(x=20, y=30)
        msg.pack(side="left", fill="x", pady=10)
        
        self.entry_var2 = tk.StringVar()
        self.input2=tk.Entry(bottomframe, textvariable=self.entry_var2)
        self.input2.place(x=115, y=30)
        self.input2.pack(side="left", fill="x", pady=10)
        #----------------------------------------------------------
        text = tk.Text(self)        
        msg = tk.Label(bottomframe1, text="Usuario:")
        msg.place(x=20, y=30)
        msg.pack(side="left", fill="x", pady=10)
        
        self.entry_var3 = tk.StringVar()
        self.input3=tk.Entry(bottomframe1, textvariable=self.entry_var3)
        self.input3.place(x=115, y=30)
        self.input3.pack(side="left", fill="x", pady=10)
        #----------------------------------------------------------
        text = tk.Text(self)        
        msg = tk.Label(bottomframe12, text="Contraseña:")
        msg.place(x=20, y=30)
        msg.pack(side="left", fill="x", pady=10)
        
        self.entry_var4 = tk.StringVar()
        self.input4=tk.Entry(bottomframe12, textvariable=self.entry_var4)
        self.input4.place(x=115, y=30)
        self.input4.pack(side="left", fill="x", pady=10)
        #----------------------------------------------------------
        
        new_button = tk.Button(bottomframe2)
        new_button ["text"] = "Nuevo"
        new_button["command"] = self.new
        
        
        new_button.pack(side="left", fill="x", pady=10)
        rec_button = tk.Button(bottomframe2)
        rec_button["text"] = "Recordar"
        rec_button["command"] = self.recovery
        rec_button.pack(side="left", fill="x", pady=10)
       
        
        
        con_button = tk.Button(bottomframe3)
        con_button ["text"] = "Conectar"
        con_button["command"] = self.get
        con_button.pack(side="left", fill="x", pady=10)
        button1 = tk.Button(bottomframe3, text="Resultados",
                            command=lambda: controller.show_frame("PageOne"))
        #button2 = tk.Button(self, text="Go to Page Two",
                            #command=lambda: controller.show_frame("PageTwo"))
            
        button1.pack(side="left", fill="x", pady=10)
        #button2.pack()
        quit = tk.Button(bottomframe3, text="QUIT", fg="red",
                              command=self.master.destroy)
        quit.pack(side="bottom", fill="x", pady=10)
        quit.place(x=140, y=180)
    def recovery(self):
        self.input1.delete(0, 'end')
        self.input2.delete(0, 'end')
        self.input3.delete(0, 'end') 
        self.input4.delete(0, 'end')
        self.input1.insert(0, "Inar")
        self.input2.insert(0, "AdventureWorks2014") 
        self.input3.insert(0, "sql_admin") 
        self.input4.insert(0, "admin")
    def new(self):
        self.input1.delete(0, 'end')
        self.input2.delete(0, 'end')
    def get(self):
        global text_sql
        global cursor
        global conn
        global con
        con=['Driver={ODBC Driver 17 for SQL Server};'
                      'Server='+str(self.entry_var1.get())+';'
                      'Database='+str(self.entry_var2.get())+';'
                      'UID='+str(self.entry_var3.get())+';'
                      'PWD='+str(self.entry_var4.get())+';'                        
                      'Trusted_Connection=yes;']
        print(con)
        conn = pyodbc.connect(con[0])

        cursor = conn.cursor()
        cursor.execute("""SELECT S.name as 'Schema',
        T.name as 'Table',
        I.name as 'Index',
        DDIPS.avg_fragmentation_in_percent,
        DDIPS.page_count
        FROM sys.dm_db_index_physical_stats (DB_ID(), NULL, NULL, NULL, NULL) AS DDIPS
        INNER JOIN sys.tables T on T.object_id = DDIPS.object_id
        INNER JOIN sys.schemas S on T.schema_id = S.schema_id
        INNER JOIN sys.indexes I ON I.object_id = DDIPS.object_id
        AND DDIPS.index_id = I.index_id
        WHERE DDIPS.database_id = DB_ID()
        and I.name is not null
        AND DDIPS.avg_fragmentation_in_percent > 0
        ORDER BY DDIPS.avg_fragmentation_in_percent desc""")
        
        for row in cursor:
            print(row)
            #print(row[3])
            #cadena.ljust(50, "=")
            #cadena.rjust(50, "=")
            #cadena.center(50, "=")
            izq=str(row[0])+"."+str(row[1])
            cen=str(row[2])
            der=str(round(float(row[3]), 2))
            #list_sql.append(text_sql+izq.ljust(40, " ")+cen.center(40, " ")+der.rjust(40, " ")+"\n")
            list_sql.append([row[0],row[1],row[2],row[3]])
            
            text_sql=text_sql+izq.ljust(55, " ")+cen.center(65, " ")+der.rjust(40, " ")+"\n"
        #conn.close()
        
class VerticalScrolledFrame(tk.Frame):
    """A pure Tkinter scrollable frame that actually works!

    * Use the 'interior' attribute to place widgets inside the scrollable frame
    * Construct and pack/place/grid normally
    * This frame only allows vertical scrolling
    """
    def __init__(self, parent, *args, **kw):
        tk.Frame.__init__(self, parent, *args, **kw)            

        # create a canvas object and a vertical scrollbar for scrolling it
        vscrollbar = tk.Scrollbar(self, orient=tk.VERTICAL)
        vscrollbar.pack(fill=tk.Y, side=tk.RIGHT, expand=tk.FALSE)
        canvas = tk.Canvas(self, bd=0, highlightthickness=0,
                        yscrollcommand=vscrollbar.set)
        canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=tk.TRUE)
        vscrollbar.config(command=canvas.yview)

        # reset the view
        canvas.xview_moveto(0)
        canvas.yview_moveto(0)

        # create a frame inside the canvas which will be scrolled with it
        self.interior = interior = tk.Frame(canvas)
        interior_id = canvas.create_window(0, 0, window=interior,
                                           anchor=tk.NW)

        # track changes to the canvas and frame width and sync them,
        # also updating the scrollbar
        def _configure_interior(event):
            # update the scrollbars to match the size of the inner frame
            size = (interior.winfo_reqwidth(), interior.winfo_reqheight())
            canvas.config(scrollregion="0 0 %s %s" % size)
            if interior.winfo_reqwidth() != canvas.winfo_width():
                # update the canvas's width to fit the inner frame
                canvas.config(width=interior.winfo_reqwidth())

        interior.bind('<Configure>', _configure_interior)

        def _configure_canvas(event):
            if interior.winfo_reqwidth() != canvas.winfo_width():
                # update the inner frame's width to fill the canvas
                canvas.itemconfigure(interior_id, width=canvas.winfo_width())
        canvas.bind('<Configure>', _configure_canvas)        




class PageOne(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        bottomframe = tk.Frame(self,parent)
        bottomframe.pack(side="top" )
        bottomframe2 = tk.Frame(bottomframe,parent)
        bottomframe2.pack(side="bottom" )
        label = tk.Label(bottomframe, text="Fragmentación", font=controller.title_font)
        label.pack(side="top", fill="x", pady=10)
        button = tk.Button(bottomframe, text="Volver atras",
                           command=lambda: controller.show_frame("StartPage"))
        button.pack()
        button2 = tk.Button(bottomframe, text="Machine learning",
                            command=lambda: controller.show_frame("PageTwo"))
        button2.pack()
        #--------
        new_button = tk.Button(bottomframe)
        new_button ["text"] = "Ver resultados"
        new_button["command"] = self.ver
        
        
        new_button.pack(side="bottom", fill="x", pady=10)
        
        #---------------------------------
        # create the text widget
        # apply the grid layout
        bottomframe2.grid_columnconfigure(0, weight=1)
        bottomframe2.grid_rowconfigure(0, weight=1)
        self.text = tk.Text(bottomframe2, height=30,width=200)
        self.text.grid(row=0, column=0, sticky='ew')
        

        # create a scrollbar widget and set its command to the text widget
        scrollbar = ttk.Scrollbar(bottomframe2, orient='vertical', command=self.text.yview)
        scrollbar.grid(row=0, column=1, sticky='ns')

        #  communicate back to the scrollbar
        self.text['yscrollcommand'] = scrollbar.set
        
    def ver(self):
        self.text.delete('1.0', 'end')
        self.text.insert(tk.INSERT, text_sql)
            


class PageTwo(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        
        self.controller = controller
        label = tk.Label(self, text="Resultados machine learning", font=controller.title_font)
        label.pack(side="top", fill="x", pady=10)
        button = tk.Button(self, text="Volver al inicio",
                           command=lambda: controller.show_frame("StartPage"))
        button.pack()
        bottomframe = tk.Frame(self,parent)
        bottomframe.pack(side="top" )
        bottomframe2 = tk.Frame(bottomframe,parent)
        bottomframe2.pack(side="bottom" )
        #label = tk.Label(bottomframe, text="Resultados machine learning", font=controller.title_font)
        #label.pack(side="top", fill="x", pady=10)
        button = tk.Button(bottomframe, text="Volver atras",
                           command=lambda: [controller.show_frame("PageOne"),reset_server(new_indice)])
        button.pack()
        #button2 = tk.Button(bottomframe, text="Aplicar machine learning",
                           # command=lambda: controller.show_frame("PageTwo"))
        #button2.pack()
        #--------
        new_button = tk.Button(bottomframe,text="Ver resultados",command=lambda list_sql=list_sql: ver2(list_sql))
        #new_button ["text"] = "Aplicar machine learning"
        #new_button["command"] = self.ver2        
        new_button.pack(side="bottom", fill="x", pady=10)
        
        
        new_button2 = tk.Button(bottomframe)
        new_button2 ["text"] = "Corregir todo"
        new_button2["command"] = self.ver
        
        
        new_button2.pack(side="bottom", fill="x", pady=10)
        
        #---------------------------------
        scframe = VerticalScrolledFrame(bottomframe2)
        scframe.pack()
        #lis = list('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
        #lis=list_sql
        #print(list_sql)
        def reset_server(i):
            
            global text_sql
            global cursor
            global conn
            global con
            text_sql=""
            
            print(con)
            conn = pyodbc.connect(con[0])

            cursor = conn.cursor()
            cursor.execute("""SELECT S.name as 'Schema',
            T.name as 'Table',
            I.name as 'Index',
            DDIPS.avg_fragmentation_in_percent,
            DDIPS.page_count
            FROM sys.dm_db_index_physical_stats (DB_ID(), NULL, NULL, NULL, NULL) AS DDIPS
            INNER JOIN sys.tables T on T.object_id = DDIPS.object_id
            INNER JOIN sys.schemas S on T.schema_id = S.schema_id
            INNER JOIN sys.indexes I ON I.object_id = DDIPS.object_id
            AND DDIPS.index_id = I.index_id
            WHERE DDIPS.database_id = DB_ID()
            and I.name is not null
            AND DDIPS.avg_fragmentation_in_percent > 0
            ORDER BY DDIPS.avg_fragmentation_in_percent desc""")
            
            for row in cursor:
                
                print(row)
                #print(row[3])
                #cadena.ljust(50, "=")
                #cadena.rjust(50, "=")
                #cadena.center(50, "=")
                izq=str(row[0])+"."+str(row[1])
                cen=str(row[2])
                der=str(round(float(row[3]), 2))
                #list_sql.append(text_sql+izq.ljust(40, " ")+cen.center(40, " ")+der.rjust(40, " ")+"\n")
                list_sql.append([row[0],row[1],row[2],row[3]])
                if i==str(row[2]):
                    text_sql=text_sql+izq.ljust(55, " ")+cen.center(65, " ")+"--nueva-fragmentacion-->>>"+der.rjust(13, " ")+"\n"
                text_sql=text_sql+izq.ljust(55, " ")+cen.center(65, " ")+der.rjust(40, " ")+"\n"
            
            
            
        def ver2(list_sql):                   
            
            #lis = list('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
            for i, x in enumerate(list_sql):
                #print(lis[i])list_sql
                if round(float(list_sql[i][3]))>20.0:                   
                
                    msg = tk.Label(scframe.interior, height=3, width=200, relief=tk.FLAT,bg="gray99",font="Dosis", text="Para la tabla ["+str(list_sql[i][0])+"."+str(list_sql[i][1])+"] con el indice ["+str(list_sql[i][2])+"] tiene un "+str(round(float(list_sql[i][3]), 2))+"% de fragmentación, se realizara: "+str(demo_machine(list_sql[i][3])))
                    msg.pack(padx=10, pady=5)
                    btn = tk.Button(scframe.interior, height=3, width=20, relief=tk.FLAT,bg="gray99", 
                        font="Dosis", text="Ejecutar corrección",
                        command=lambda i=i,x=x: openlink(i,list_sql[i][3]))
                    btn.pack(padx=10, pady=5,side=tk.TOP)
                #c1 = tk.Checkbutton(scframe.interior,variable=i, onvalue=1, offvalue=0, command=lambda i=i,x=x: openlink(i,demo_machine(list_sql[i][2]))))
                #c1.pack(padx=10, pady=5)
        def demo_machine(value):
            if value>40.0:
                return "RE-BUILD"
            elif value<40.0 :#and value>20.0:
                return "RE-ORGANIZE"
            else:
                return " "
        
        def openlink(i,value):
            global con
            global cursor
            global new_indice
            new_indice=str(list_sql[i][2])
            print(con)
            print(cursor)
            #conn = pyodbc.connect(con[0])
            #cursor = conn.cursor()
            if(messagebox.askyesno(message="¿Desea continuar?", title="Confirmación")):
                #ventana=process()
                if value>30.0:
                    
                    print( "ALTER INDEX ["+str(list_sql[i][2])+"] ON ["+str(list_sql[i][0])+"].["+str(list_sql[i][1])+"] REBUILD")
                    print("Consulta en ejecucion...")
                    startTime = time.time()
                    cursor.execute("ALTER INDEX ["+str(list_sql[i][2])+"] ON ["+str(list_sql[i][0])+"].["+str(list_sql[i][1])+"] REBUILD")
                    conn.commit()
                    endTime = time.time()
                    #startTime = time.time()
                    #for i in range(0,20):
                    #   print(i)
                       # making delay for 1 second
                    #   time.sleep(1)
                    #endTime = time.time()
                    elapsedTime = endTime - startTime
                    print("Elapsed Time = %s" % elapsedTime)
                    print("termino de la ejecucion")
                    #ventana.destroy
                    cursor.close
                    conn.close()
                    
                elif value<30.0: # and value>20.0:
                    print ("ALTER INDEX ["+str(list_sql[i][2])+"] ON ["+str(list_sql[i][0])+"].["+str(list_sql[i][1])+"]  REORGANIZE")
                    print("Consulta en ejecucion...")
                    cursor.execute( "ALTER INDEX ["+str(list_sql[i][2])+"] ON ["+str(list_sql[i][0])+"].["+str(list_sql[i][1])+"]  REORGANIZE")
                    conn.commit()
                    #time.sleep(30.0)
                    #print("Termino de la ejecucion.")
                    #ventana.destroy
                    #cursor.close
                    #conn.close()
                else:
                    return None
        def process():
            
            #win=tk.Toplevel()
            #win.geometry()
            #win.configure(background=)
            #e3=tk.Label(win,text='Procesando...')
            #e3.pack(padx=5, pady=5,ipady=5,fill=tk.X)
            #root = tk.Tk()
            root = tk.Toplevel()
            file="gif1.gif"

            info = Image.open(file)

            frames = info.n_frames  # gives total number of frames that gif contains

            # creating list of PhotoImage objects for each frames
            im = [tk.PhotoImage(file=file,format=f"gif -index {i}") for i in range(frames)]

            count = 0
            anim = None
            gif_label = tk.Label(root,image="")
            gif_label.pack()

            #start = tk.Button(root,text="start",command=lambda :animation(count,gif_label,root,im, frames))
            #start.pack()

            #stop = tk.Button(root,text="stop",command=stop_animation)
            #stop.pack()

            #root.mainloop()
            return root
        def animation(count,gif_label,root,im,frames):
            global anim
            im2 = im[count]

            gif_label.configure(image=im2)
            count += 1
            if count == frames:
                count = 0
            anim = root.after(50,lambda :animation(count))

        def stop_animation(root):
            root.after_cancel(anim)

    
        
       
    def ver(self):
        for i, x in enumerate(list_sql):
            value=demo_machine(list_sql[i][3])
            if value>30.0:
                    print( "ALTER INDEX ["+str(list_sql[i][2])+"] ON ["+str(list_sql[i][0])+"].["+str(list_sql[i][1])+"] REBUILD")
                    cursor.execute("ALTER INDEX ["+str(list_sql[i][2])+"] ON ["+str(list_sql[i][0])+"].["+str(list_sql[i][1])+"] REBUILD")
                    #ALTER INDEX indice Z
            elif value<30.0:# and value>20.0:
                print ("ALTER INDEX ["+str(list_sql[i][2])+"] ON ["+str(list_sql[i][0])+"].["+str(list_sql[i][1])+"]  REORGANIZE")
                cursor.execute( "ALTER INDEX ["+str(list_sql[i][2])+"] ON ["+str(list_sql[i][0])+"].["+str(list_sql[i][1])+"]  REORGANIZE")

            else:
                return None
    
    

    
    
                
                
            
            
             

            
            #print(list_sql[i])
if __name__ == "__main__":
    app = SampleApp()
    app.mainloop()
```
