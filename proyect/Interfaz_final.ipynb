```python
try:
    import tkinter as tk                # python 3
    from tkinter import font as tkfont  # python 3
except ImportError:
    import Tkinter as tk     # python 2
    import tkFont as tkfont  # python 2
import pyodbc
class SampleApp(tk.Tk):

    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)

        self.title_font = tkfont.Font(family='IA', size=18, weight="bold", slant="italic")
        self.configure(width=330, height=400)
        # the container is where we'll stack a bunch of frames
        # on top of each other, then the one we want visible
        # will be raised above the others
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)

        self.frames = {}
        for F in (StartPage, PageOne, PageTwo):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame

            # put all of the pages in the same location;
            # the one on the top of the stacking order
            # will be the one that is visible.
            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame("StartPage")

    def show_frame(self, page_name):
        '''Show a frame for the given page name'''
        frame = self.frames[page_name]
        frame.tkraise()


class StartPage(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        self.master = parent
        
        #self.place(relwidth=1, relheight=1)
        
        self.pack()
        self.create_widgets()
        self.pack()
        label = tk.Label(self, text="Pagina de inicio", font=controller.title_font)
        label.pack(side="top", fill="x", pady=10)
        self.pack()
        button1 = tk.Button(self, text="Resultados",
                            command=lambda: controller.show_frame("PageOne"))
        #button2 = tk.Button(self, text="Go to Page Two",
                            #command=lambda: controller.show_frame("PageTwo"))
        button1.pack()
        #button2.pack()
    def create_widgets(self):
        self.pack()
        self.text = tk.Text(self)        
        self.msg = tk.Label(self.master, text="Nombre servidor")
        self.msg.place(x=20, y=10)
       
        self.entry_var = tk.StringVar()
        self.input=tk.Entry(self, textvariable=self.entry_var)
        self.input.place(x=115, y=10)
        #-----------------------------------------------------------
        
        
        self.pack()
        self.text = tk.Text(self)        
        self.msg = tk.Label(self.master, text="Nombre DB")
        self.msg.place(x=20, y=30)
        self.pack()
        self.pack()
        self.entry_var = tk.StringVar()
        self.input2=tk.Entry(self, textvariable=self.entry_var)
        self.input2.place(x=115, y=30)
        #-----------------------------------------------------------
        
        
        self.pack()
        self.text = tk.Text(self)        
        self.msg = tk.Label(self.master, text="Usuario")
        self.msg.place(x=20, y=50)
        self.pack()
        self.pack()
        self.entry_var = tk.StringVar()
        self.input3=tk.Entry(self, textvariable=self.entry_var)
        self.input3.place(x=115, y=50)
        #-----------------------------------------------------------
         #-----------------------------------------------------------
        
        
        self.pack()
        self.text = tk.Text(self)        
        self.msg = tk.Label(self.master, text="Password")
        self.msg.place(x=20, y=70)
        self.pack()
        self.pack()
        self.entry_var = tk.StringVar()
        self.input4=tk.Entry(self, textvariable=self.entry_var)
        self.input4.place(x=115, y=70)
        #-----------------------------------------------------------
        self.pack()
        self.hi_there1 = tk.Button(self)
        self.hi_there1["text"] = "Nuevo"
        self.hi_there1["command"] = self.new
        self.hi_there1.pack(side="top")
        self.hi_there1.place(x=100, y=100)
        self.pack()
        self.hi_there2 = tk.Button(self)
        self.hi_there2["text"] = "Recordar"
        self.hi_there2["command"] = self.recovery
        self.hi_there2.pack(side="top")
        self.hi_there2.place(x=170, y=100)
        
        
        
        
        
        
        
        self.hi_there3 = tk.Button(self)
        self.hi_there3["text"] = "Conectar"
        self.hi_there3["command"] = self.get
        self.hi_there3.pack(side="top")
        self.hi_there3.place(x=145, y=140)
        
        
        
        self.quit = tk.Button(self, text="QUIT", fg="red",
                              command=self.master.destroy)
        self.quit.pack(side="bottom")
        self.quit.place(x=140, y=180)
        
        

    def get(self):
         
        conn = pyodbc.connect('Driver={ODBC Driver 17 for SQL Server};'
                              'Server=Inar;'
                              'Database=AdventureWorks2014;'
                              'Trusted_Connection=yes;')

        cursor = conn.cursor()
        cursor.execute("""SELECT  OBJECT_NAME(IDX.OBJECT_ID) AS Table_Name, 
        IDX.name AS Index_Name, 
        IDXPS.index_type_desc AS Index_Type, 
        IDXPS.avg_fragmentation_in_percent  Fragmentation_Percentage
        FROM sys.dm_db_index_physical_stats(DB_ID(), NULL, NULL, NULL, NULL) IDXPS 
        INNER JOIN sys.indexes IDX  ON IDX.object_id = IDXPS.object_id 
        AND IDX.index_id = IDXPS.index_id 
        ORDER BY Fragmentation_Percentage DESC""")
        text_sql=""
        for row in cursor:
            #print(row)
            print(row[3])
            text_sql=text_sql+str(row[3])+"\n"
        
        
    
    def recovery(self):
        self.input.delete(0, 'end')
        self.input2.delete(0, 'end')
        self.input3.delete(0, 'end') 
        self.input4.delete(0, 'end')
        self.input.insert(0, "Inar")
        self.input2.insert(0, "AdventureWorks2014") 
        self.input3.insert(0, "admin") 
        self.input4.insert(0, "admin")
    def new(self):
        self.input.delete(0, 'end')
        self.input2.delete(0, 'end')
        self.input3.delete(0, 'end') 
        self.input4.delete(0, 'end')


class PageOne(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        label = tk.Label(self, text="This is page 1", font=controller.title_font)
        label.pack(side="top", fill="x", pady=10)
        button = tk.Button(self, text="Go to the start page",
                           command=lambda: controller.show_frame("StartPage"))
        button.pack()


class PageTwo(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        label = tk.Label(self, text="This is page 2", font=controller.title_font)
        label.pack(side="top", fill="x", pady=10)
        button = tk.Button(self, text="Go to the start page",
                           command=lambda: controller.show_frame("StartPage"))
        button.pack()
if __name__ == "__main__":
    app = SampleApp()
    app.mainloop()
```
